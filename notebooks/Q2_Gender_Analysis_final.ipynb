```python
# prompt: Use the above code cell and plot a similar graph with youth olympics removed and take out the olympics in x axis and keep the year

import pandas as pd
import matplotlib.pyplot as plt

# Assuming winter_df is your DataFrame and it has columns 'game', 'gender', and 'id'

# Remove rows containing "Youth" in the 'game' column
winter_df_no_youth = winter_df[~winter_df['game'].str.contains('Youth')]

# Extract the year from the 'game' column
winter_df_no_youth['Year'] = winter_df_no_youth['game'].str.extract(r'(\d{4})')

# Group by year and gender, and count the number of athletes
gender_year_counts = winter_df_no_youth.groupby(['Year', 'gender'])['id'].count().unstack()

# Create a stacked bar chart
ax = gender_year_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

# Customize the chart
plt.title('Athlete Distribution by Gender in Winter Olympics Over Years (Youth Olympics Removed)')
plt.xlabel('Year')
plt.ylabel('Number of Athletes')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Gender')

# Show the chart
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

```python
# prompt: use the above code cell and do it for the summer olympics for as many years that can fit on the graph

import pandas as pd
import matplotlib.pyplot as plt
# Assuming summer_df is your DataFrame and it has columns 'game', 'gender', and 'id'

# Remove rows containing "Youth" in the 'game' column
summer_df_no_youth = summer_df[~summer_df['game'].str.contains('Youth')]

# Extract the year from the 'game' column
summer_df_no_youth['Year'] = summer_df_no_youth['game'].str.extract(r'(\d{4})')

# Group by year and gender, and count the number of athletes
gender_year_counts = summer_df_no_youth.groupby(['Year', 'gender'])['id'].count().unstack()

# Create a stacked bar chart
ax = gender_year_counts.plot(kind='bar', stacked=True, figsize=(15, 8))  # Increase figure size for more years

# Customize the chart
plt.title('Athlete Distribution by Gender in Summer Olympics Over Years (Youth Olympics Removed)')
plt.xlabel('Year')
plt.ylabel('Number of Athletes')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Gender')

# Show the chart
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

```python
# prompt: include percentages of male to female in the chart representation above

import pandas as pd
import matplotlib.pyplot as plt

# Assuming summer_df_no_youth is your DataFrame and it has columns 'Year', 'gender', and 'id'

# Group by year and gender, and count the number of athletes
gender_year_counts = summer_df_no_youth.groupby(['Year', 'gender'])['id'].count().unstack()

# Calculate the total number of athletes for each year
gender_year_counts['Total'] = gender_year_counts['Male'] + gender_year_counts['Female']

# Calculate the percentage of male and female athletes for each year
gender_year_counts['Male Percentage'] = (gender_year_counts['Male'] / gender_year_counts['Total']) * 100
gender_year_counts['Female Percentage'] = (gender_year_counts['Female'] / gender_year_counts['Total']) * 100

# Create a stacked bar chart
ax = gender_year_counts[['Male', 'Female']].plot(kind='bar', stacked=True, figsize=(15, 8))

# Customize the chart
plt.title('Athlete Distribution by Gender in Summer Olympics Over Years (Youth Olympics Removed)')
plt.xlabel('Year')
plt.ylabel('Number of Athletes')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Gender')

# Add percentage labels to the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.0f}', ha='center', va='center')

# Show the chart
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()

# Create a line chart for percentage change
ax2 = gender_year_counts[['Male Percentage', 'Female Percentage']].plot(kind='line', figsize=(15, 6))
plt.title('Proportion of Male to Female Athletes in Summer Olympics Over Years')
plt.xlabel('Year')
plt.ylabel('Percentage of Athletes')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()

```

```python
# prompt: compare number of men to females based on country type and year of olympics, make a stacked bar chart

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'athlete_df' is your DataFrame with 'gender', 'country_code', and 'year' columns

# Group by country and year, then count the number of male and female athletes
gender_counts = athlete_df.groupby(['team', 'game', 'gender'])['id'].count().unstack()

# Reset the index to make 'country_code' and 'year' regular columns
gender_counts = gender_counts.reset_index()

# Fill NaN values with 0 (if there are no athletes of a certain gender in a country/year)
gender_counts = gender_counts.fillna(0)

# Create a stacked bar chart
fig, ax = plt.subplots(figsize=(15, 8))

# Specify the width of the bars
width = 0.5

# Create the stacked bar chart
# Use 'game' for the x-axis and the correct column names for 'Male' and 'Female'
ax.bar(gender_counts['game'], gender_counts['Male'], width, label='Male', color='blue') # Changed 'M' to 'Male'
ax.bar(gender_counts['game'], gender_counts['Female'], width, bottom=gender_counts['Male'], label='Female', color='red') # Changed 'F' to 'Female'

# Customize the plot
ax.set_xlabel('Year')
ax.set_ylabel('Number of Athletes')
ax.set_title('Comparison of Male and Female Athletes by Country and Year')
ax.legend()

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
